## __Pandas의 활용__
- csv 파일을 pandas를 통해 분석하고자 함.
- __view__ attribute의 의미
    1. AP, PA는 chest에서의 어떤 질병을 나타내는 지표 인듯함.

```python
# 메타데이터 csv파일 pandas로 불러오기
df = pd.read_csv('./covid-chestxray-dataset/metadata.csv')

# finding attribute가 "COVID-19"인 사람만 불러오기
selected_df = df[df.finding=="COVID-19"]

# 이들 중 view가 "AP"이거나 "PA"인 사람 불러오기
selected_df = selected_df[(selected_df.view == "AP") | (selected_df.view == "PA")]

# 이중 상위 10개의 데이터 구성 살펴보기
selected_df.head(10)
```

```python
# selected_df의 key를 모두 살펴보기.
selected_df.keys()
```

```python
# selected_df.filename.values

# 여기 metadata 중에서 filename부분 불러오기
# 이 때, selected_df.filename로 부르면 여전히 pd 파일이기에 value들만을 받고자함.
# 이 경우 values를 쓰면 array로 받아서 다시 tolist()를 통해 list형태로 타입을 바꿈.
images = selected_df.filename.values.tolist()
```

### 커스텀 데이터 만들기 1단계
- 오픈 데이터 1 중 covid19 사진을 우리가 지정하고자하는 폴더 디렉토리로 이동하고자 함.

```python
# 경로 지정해주기
path = './COVID19-DATASET/train/'
train_list = ['covid19', 'normal']

# './COVID19-DATASET/train/covid19'와 './COVID19-DATASET/train/normal'의 폴더를 만들고자 함.
# 이를 조금 더 정돈된 형태의 코딩을 위해 for문으로 아래처럼 정리함.
for tmp in train_list:
    tmp_path = os.path.join(path, tmp)
    if not os.path.isdir(tmp_path):
        os.makedirs(tmp_path)
```

```python
# images는 './covid-chestxray-dataset'의 디렉토리의 covid인 사람의 데이터
for image in images:
    shutil.copy(os.path.join('./covid-chestxray-dataset/images', image), os.path.join(path + train_list[0], image))
```

### 커스텀 데이터 만들기 2단계
- 오픈 데이터 2 중 normal 사진을 우리가 지정하고자하는 폴더 디렉토리로 이동하고자 함.

```python
# 이는 다른 저장소의 chest가 Normal인 사람들의 사진들
for image in os.listdir('./chest_xray/train/NORMAL')[:300]:
    shutil.copy(os.path.join('./chest_xray/train/NORMAL', image), os.path.join(path + train_list[1], image))
```

```python
DATA_PATH = './COVID19-DATASET/train'
```

```python
# covid19와 normal의 데이터의 클래스 개수의 비율을 보기 위한 함수
def show_classes(DATA_PATH):
    # DATA_PATH: data가 들어있는 경로
    # 'covid19'와 'normal'에 대한 path를 받음.
    class_names = os.listdir(DATA_PATH)
    image_count = {}

    # 해당 path로 접근해 데이터의 개수를 받아옴.
    for i in class_names:
        image_count[i] = len(os.listdir(os.path.join(DATA_PATH,i)))

    #Plotting Distribution of Each Classes
    fig1, ax1 = plt.subplots()
    ax1.pie(image_count.values(),
            labels = image_count.keys(),
            shadow=True,
            autopct = '%1.1f%%',
            startangle=90)
    plt.show()
```

```python
# covid19와 normal의 각 이미지의 height과 width가 어떻게 달라지는지를 보기 위한 함수
def get_stat(DATA_PATH):
    class_names = os.listdir(DATA_PATH)
    
    height = {}
    width = {}

    for clss in class_names:
        tmp_path = os.path.join(DATA_PATH, clss)
        images = os.listdir(tmp_path)

        height[clss] = []
        width[clss] = []

        for image in images:
            img = cv2.imread(os.path.join(tmp_path, image))

            height[clss].append(img.shape[0])
            width[clss].append(img.shape[1])
            
    return height, width
```

```python
show_classes(DATA_PATH) # DATA_PATH의 train data의 클래스별 비율
```

```python
height, width = get_stat(DATA_PATH)
```

```python
plt.hist(height['normal'], bins=50, label = 'normal: height')
plt.hist(width['normal'], bins=50, label = 'normal: width')
plt.legend()
```

```python
plt.hist(height['covid19'], bins=50, label = 'normal: height')
plt.hist(width['covid19'], bins=50, label = 'normal: width')
plt.legend()
```

## __사진 확인하기__
- plt와 cv2를 이용해, image 확인해보기

```python
fig = plt.figure(figsize=(16,5))
fig.suptitle("COVID19 Positive", size=22)
img_paths = os.listdir(path + 'covid19')
shuffle(img_paths)

for i,image in enumerate(img_paths[:4]):
    img = cv2.imread(os.path.join(path + 'covid19', image))
    plt.subplot(1,4, i+1, frameon=False)
    plt.imshow(img)
fig.show()
```

```python
fig = plt.figure(figsize=(16,5))
fig.suptitle("COVID19 Negative - Healthy", size=22)
img_paths = os.listdir(path + 'normal')
shuffle(img_paths)

for i,image in enumerate(img_paths[:4]):
    img = cv2.imread(os.path.join(path + 'normal', image))
    plt.subplot(1,4, i+1, frameon=False)
    plt.imshow(img)
fig.show()
```
